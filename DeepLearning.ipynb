```python
def criar_rede():
    k.clear_session()
    rede_neural = Sequential([
        # entrada
        tf.keras.layers.InputLayer(shape=(30,)),

        # camada oculta 1 e 2
        tf.keras.layers.Dense(units=16, activation='relu', # units = (numero de entradas (as 30 colunas) + numero de neuronios na camada de saida (1 pois esse problema é classificação... ou o tumor é bom ou ruim)) / 2
                            kernel_initializer='random_uniform'),

        tf.keras.layers.Dropout(rate=0.2), # apagando 20% da camada oculta para não ter overfitting

        tf.keras.layers.Dense(units=16, activation='relu', # units = (numero de entradas (as 30 colunas) + numero de neuronios na camada de saida (1 pois esse problema é classificação... ou o tumor é bom ou ruim)) / 2
                            kernel_initializer='random_uniform'),

        tf.keras.layers.Dropout(rate=0.2), # apagando 20% da camada oculta para não ter overfitting
        
        # saída
        tf.keras.layers.Dense(units=1, activation='sigmoid')])
    otimazador = tf.keras.optimizers.Adam(learning_rate=0.001, clipvalue=0.5)
    rede_neural.compile(optimizer=otimazador, loss='binary_crossentropy', metrics=['binary_accuracy'])
    return rede_neural
```

```python
base = pd.read_csv("C:/Users/danil/OneDrive/Documentos/Deep Learning com Python de A a Z-20240819T113932Z-001/Deep Learning com Python de A a Z/Parte 1 - Redes Neurais Artificiais/classificação mais classes/iris.csv")
```

```python
x = base.iloc[:, 0:4].values
y = base.iloc[:, 4].values
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)
y = np_utils.to_categorical(y)
```

```python
rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape=(4,)),
    tf.keras.layers.Dense(units = 8, activation = 'relu', 
                            kernel_initializer = 'normal', input_dim = 4),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(units = 8, activation = 'relu', 
                            kernel_initializer = 'normal'),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(units = 3, activation = 'softmax')])

rede_neural.compile(optimizer = 'adam', 
                      loss = 'categorical_crossentropy', 
                      metrics = ['categorical_accuracy'])
rede_neural.fit(x, y, 
                  batch_size = 10, epochs = 2000)
```

```python
rede_neural.evaluate(x, y)
```

```python
rede_neural.save('rede_neural_iris.keras')
```

### exercício tuning automoveis

```python
import pandas as pd
import tensorflow as tf
import scikeras
import sklearn
from tensorflow.keras import backend as k
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
base = pd.read_csv('C:/Users/danil/OneDrive/Documentos/Deep Learning com Python de A a Z-20240819T113932Z-001/Deep Learning com Python de A a Z/Parte 1 - Redes Neurais Artificiais/regressão um valor/autos.csv', encoding='ISO-8859-1')
```

```python
base = base.drop('dateCrawled', axis=1)
base = base.drop('dateCreated', axis=1)
base = base.drop('nrOfPictures', axis=1)
base = base.drop('postalCode', axis=1)
base = base.drop('lastSeen', axis=1)
base = base.drop('offerType', axis=1)
base = base.drop('seller', axis=1)
base = base.drop('name', axis=1)
```

```python
base = base[base.price > 10]
base = base.loc[base.price < 350000]
```

```python
valores = {
    'vehicleType' : 'limousine',
    'gearbox' : 'manuell',
    'model' : 'golf',
    'fuelType' : 'benzin',
    'notRepairedDamage' : 'nein'
}
base = base.fillna(value=valores)
```

```python
x = base.iloc[:, 1: 12].values
y = base.iloc[:, 0].values
```

```python
labelencoder_previsores = LabelEncoder()
x[:, 0] = labelencoder_previsores.fit_transform(x[:, 0])
x[:, 1] = labelencoder_previsores.fit_transform(x[:, 1])
x[:, 3] = labelencoder_previsores.fit_transform(x[:, 3])
x[:, 5] = labelencoder_previsores.fit_transform(x[:, 5])
x[:, 8] = labelencoder_previsores.fit_transform(x[:, 8])
x[:, 9] = labelencoder_previsores.fit_transform(x[:, 9])
x[:, 10] = labelencoder_previsores.fit_transform(x[:, 10])
```

```python
onehotencoder = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [0, 1, 3, 5, 8, 9, 10])], remainder='passthrough')
x = onehotencoder.fit_transform(x).toarray()
```

```python
def criar_rede(loss):
    k.clear_session()
    regressor = Sequential([
        tf.keras.layers.InputLayer(shape=(316,)),
        tf.keras.layers.Dense(units=158, activation='relu', kernel_initializer='random_uniform'),
        tf.keras.layers.Dropout(rate=0.2),
        tf.keras.layers.Dense(units=158, activation='relu', kernel_initializer='random_uniform'),
        tf.keras.layers.Dropout(rate=0.2),
        tf.keras.layers.Dense(units=1, activation='linear')])
    regressor.compile(loss=loss, optimizer='adam', metrics=['mean_absolute_error'])
    return regressor
```

```python
regressor = KerasRegressor(model=criar_rede, epochs=100, batch_size=300)
```

```python
parametros = {
    'model__loss': ['mean_squared_error', 'mean_absolute_error',
                       'mean_absolute_percentage_error', 
                       'mean_squared_logarithmic_error',
                       'squared_hinge']
            } # squared_hinge foi o melhor
```

```python
grid_search = GridSearchCV(estimator=regressor, param_grid=parametros, cv=10)
```

```python
grid_search = grid_search.fit(x, y)
```

```python
melhores_parametros = grid_search.best_params_
```
