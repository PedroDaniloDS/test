## breast cancer cruzada

```python
import pandas as pd
import tensorflow as tf
import sklearn
import scikeras
from sklearn.model_selection import cross_val_score
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as k
```

```python
pd.__version__, tf.__version__, sklearn.__version__, scikeras.__version__
```

```python
x = pd.read_csv("C:/Users/danil/Downloads/entradas_breast.csv")
y = pd.read_csv("C:/Users/danil/Downloads/saidas_breast.csv")
```

```python
def criar_rede():
    k.clear_session()
    rede_neural = Sequential([
        # entrada
        tf.keras.layers.InputLayer(shape=(30,)),

        # camada oculta 1 e 2
        tf.keras.layers.Dense(units=16, activation='relu', # units = (numero de entradas (as 30 colunas) + numero de neuronios na camada de saida (1 pois esse problema é classificação... ou o tumor é bom ou ruim)) / 2
                            kernel_initializer='random_uniform'),

        tf.keras.layers.Dropout(rate=0.2), # apagando 20% da camada oculta para não ter overfitting

        tf.keras.layers.Dense(units=16, activation='relu', # units = (numero de entradas (as 30 colunas) + numero de neuronios na camada de saida (1 pois esse problema é classificação... ou o tumor é bom ou ruim)) / 2
                            kernel_initializer='random_uniform'),

        tf.keras.layers.Dropout(rate=0.2), # apagando 20% da camada oculta para não ter overfitting
        
        # saída
        tf.keras.layers.Dense(units=1, activation='sigmoid')])
    otimazador = tf.keras.optimizers.Adam(learning_rate=0.001, clipvalue=0.5)
    rede_neural.compile(optimizer=otimazador, loss='binary_crossentropy', metrics=['binary_accuracy'])
    return rede_neural
```

```python
rede_neural = KerasClassifier(model= criar_rede, 
                              epochs=100, 
                              batch_size=10)
```

```python
resultados = cross_val_score(estimator=rede_neural,
                             X=x, y=y,
                             cv=10, scoring='accuracy')
```
